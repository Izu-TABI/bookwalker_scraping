```python
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd
import os
import datetime
```

```python
url = "https://bookwalker.jp/category/3/?qspp=1&np=1"
response = req.urlopen(url)
```

```python
parse_html = BeautifulSoup(response, 'html.parser')
```

```python
book_lists = parse_html
```

```python
book_title = book_lists.find_all('a','m-book-item__title')
book_price = book_lists.find_all('span','m-book-item__price-num')
book_title[1:10]
book_price[1:10]
```

```python
book_lists = []
price_list = []

for i in book_title :
    book_lists.append(i.attrs['title'])

for i in book_price :
    price_list.append(i.string)
    
```

```python
df_books = pd.DataFrame({'【期間限定価格】ライトノベル作品':book_lists, '値段':price_list})
```

```python
df_notnull = df_books.dropna(how='any')
```
